```python
import os
import yaml
import glob
import cv2
import matplotlib.pyplot as plt
from ultralytics import YOLO
from IPython.display import Image, display

# --- CU HNH ---
# Hy sa ng dn ny thnh ni bn va upload file best.pt ln
# Nu upload trc tip, thng n nm ngay  th mc hin ti hoc /kaggle/working/
MODEL_PATH = "/kaggle/input/result-conv-s/best.pt" 

# ng dn tp test (gi nguyn theo dataset ca bn)
TEST_IMAGES_DIR = "/kaggle/input/hituav-a-highaltitude-infrared-thermal-dataset/hit-uav/images/test"

# ======================================================================================
# 1. TO LI FILE DATASET.YAML (V session mi nn cn to li  model bit ng dn)
# ======================================================================================
data_dict = {
    "train": "/kaggle/input/hituav-a-highaltitude-infrared-thermal-dataset/hit-uav/images/train",
    "val": "/kaggle/input/hituav-a-highaltitude-infrared-thermal-dataset/hit-uav/images/val",
    "test": "/kaggle/input/hituav-a-highaltitude-infrared-thermal-dataset/hit-uav/images/test",
    "nc": 5,
    "names": ['Person', 'Car', 'Bicycle', 'OtherVehicle', 'DontCare']
}
with open("dataset.yaml", "w") as f:
    yaml.dump(data_dict, f, sort_keys=False)
print("  to li file dataset.yaml")

# ======================================================================================
# 2. LOAD MODEL V NH GI (Validation)
# ======================================================================================
if os.path.exists(MODEL_PATH):
    print(f" ang load model t {MODEL_PATH}...")
    model = YOLO(MODEL_PATH)
    
    # Chy nh gi trn tp TEST
    print(" ang tnh ton cc ch s trn tp Test (vic ny mt vi pht)...")
    metrics = model.val(data="dataset.yaml", split='test', project='runs/val', name='result_view')
    
    print("\n" + "="*40)
    print(f" KT QU NH GI :")
    print(f"   - mAP50        : {metrics.box.map50:.4f}")
    print(f"   - mAP50-95     : {metrics.box.map:.4f}")
    print(f"   - Precision    : {metrics.box.mp:.4f}")
    print(f"   - Recall       : {metrics.box.mr:.4f}")
    print("="*40 + "\n")
    
    # Hin th Confusion Matrix (Ma trn nhm ln)
    # YOLO t ng lu file ny khi chy val()
    confusion_matrix_path = 'runs/val/result_view/confusion_matrix.png'
    if os.path.exists(confusion_matrix_path):
        print(" MA TRN NHM LN (Confusion Matrix):")
        display(Image(filename=confusion_matrix_path, width=600))
    
    # Hin th biu  F1-Score (nu c)
    f1_curve_path = 'runs/val/result_view/F1_curve.png'
    if os.path.exists(f1_curve_path):
        print(" BIU  F1-SCORE:")
        display(Image(filename=f1_curve_path, width=600))

else:
    print(f" Li: Khng tm thy file '{MODEL_PATH}'. Hy kim tra li bn  upload cha?")

```

```python
# ======================================================================================
# 3. D ON V HIN TH NH (Inference) -  FIX LI TM FILE
# ======================================================================================
print("\n ang chy th nghim trn mt s nh Test...")

# Chy predict (Gi nguyn)
results = model.predict(source=TEST_IMAGES_DIR, max_det=10, conf=0.25, save=True, project='runs/predict', name='demo', imgsz=640)

# Tm th mc kt qu va lu
predict_dirs = glob.glob('runs/predict/demo*')

if predict_dirs:
    latest_pred_dir = max(predict_dirs, key=os.path.getctime)
    
    # --- SA  Y: Dng recursive=True  tm trong c th mc con ---
    # Tm tt c file .jpg trong mi ngc ngch ca folder kt qu
    pred_images = glob.glob(os.path.join(latest_pred_dir, '**', '*.jpg'), recursive=True)
    
    # Nu khng thy jpg th tm th png (phng h)
    if not pred_images:
        pred_images = glob.glob(os.path.join(latest_pred_dir, '**', '*.png'), recursive=True)

    print(f" Tm thy {len(pred_images)} nh kt qu.")

    # V 6 nh ln li
    if len(pred_images) > 0:
        plt.figure(figsize=(15, 10))
        # Ch ly 6 nh u tin tm c
        for i, img_path in enumerate(pred_images[:6]): 
            img = cv2.imread(img_path)
            if img is None: continue # B qua nu li c file
            
            img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
            plt.subplot(2, 3, i + 1)
            plt.imshow(img)
            plt.axis('off')
            # Ly tn file cho ngn gn
            filename = os.path.basename(img_path)
            plt.title(f"Result: {filename}", fontsize=10)
            
        plt.tight_layout()
        plt.show()
    else:
        print(" Vn khng tm thy nh no. Hy kim tra li folder runs/predict.")
else:
    print(" Khng tm thy th mc output no.")
```
